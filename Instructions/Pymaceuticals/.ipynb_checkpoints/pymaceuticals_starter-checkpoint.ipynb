## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')

combined_data_df
```

```python
# Checking the number of mice.
number_of_mice = combined_data_df["Mouse ID"].nunique()
number_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse_id = combined_data_df[combined_data_df.duplicated(["Mouse ID", "Timepoint"])]
duplicate_mouse_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_id_all = combined_data_df[combined_data_df.duplicated(["Mouse ID", "Timepoint"], keep=False)]
duplicate_mouse_id_all
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
 
mouse_df  = combined_data_df[ combined_data_df['Mouse ID'] == "g989" ].index
  
combined_data_df.drop(mouse_df, inplace = True)

combined_data_df
```

```python
# Checking the number of mice in the clean DataFrame.
num_mice = combined_data_df["Mouse ID"].nunique()
num_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drug_group_data = combined_data_df.groupby(["Drug Regimen"])

drug_group_mean = drug_group_data.mean()
drug_group_median = drug_group_data.median()
drug_group_var = drug_group_data.var()
drug_group_std = drug_group_data.std()
drug_group_sem = drug_group_data.sem()


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drug_group_mean
drug_group_median
drug_group_var
drug_group_std 
drug_group_sem
# Using the aggregation method, produce the same summary statistics in a single line
drug_group_data.agg(['mean', 'median', 'var', 'std', 'sem'])
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pandas.

grouped_mice_data = combined_data_df.groupby("Drug Regimen")
grouped_mice_data = grouped_mice_data.agg({"Mouse ID":"nunique"})
grouped_mice_data = grouped_mice_data.reset_index()

mice_regimen_data = grouped_mice_data.set_index("Drug Regimen")

mice_regimen_data.plot(kind="bar", figsize=(10,10), color='b', alpha = .75, align="center")
plt.title("Number of Mice Per Drug Regimen")
plt.xlabel("Drug Regimen ")
plt.ylabel("Number of Mice")
plt.ylim(0,30)
plt.show()
plt.tight_layout()
```

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pyplot.
mice = mice_regimen_data["Mouse ID"]
x_axis = np.arange(len(mice_regimen_data))

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations,["Capomulin", "Ceftamin", "Infubinol","Ketapril","Naftisol","Placebo","Propriva","Ramicane","Stelasyn","Zoniferol"], rotation="vertical")

plt.bar(x_axis, mice, color='b', alpha = .75, align="center")
plt.title("Number of Mice Per Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Mice")
plt.ylim(0,30)
plt.show()
plt.tight_layout()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas

combined_data_df["Sex"].value_counts().plot(kind='pie', colors = ["green", "orange"], autopct="%1.1f%%") 
plt.title("Sex of Mice")
combined_data_df["Sex"].value_counts()
plt.show()
```
